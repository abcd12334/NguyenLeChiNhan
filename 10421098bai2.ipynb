```python
import pandas as pd
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|', index_col='user_id')
print(users.head())
print(users.info())
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print(mean_age_per_occupation)
# Create a function to calculate the male ratio
def male_ratio(group):
    male_count = (group['gender'] == 'M').sum()
    total_count = group['gender'].count()
    return (male_count / total_count) * 100
male_ratio_per_occupation = users.groupby('occupation').apply(male_ratio).sort_values(ascending=False)
print(male_ratio_per_occupation)
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_age_per_occupation)
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_per_occupation_gender)
gender_counts = users.groupby(['occupation', 'gender']).size()
occupation_counts = users.groupby('occupation').size()
occupation_gender_percentages = (gender_counts / occupation_counts) * 100
print(occupation_gender_percentages)
```
